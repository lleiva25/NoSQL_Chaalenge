### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene':20}

# Use count_documents to display the number of documents in the result
hygiene_20_count = establishments.count_documents(query)
print("There are " + str(hygiene_20_count) + " documents for this query")
print("-"*50)

#Save the results of the query to variable
hygiene_20 = list(establishments.find(query))


# Display the first document in the results using pprint
pprint(hygiene_20[0])
```

```python
#Import dependances
import pandas as pd

# Convert the result to a Pandas DataFrame
hygiene_20_df = pd.DataFrame(hygiene_20)

# Display the number of rows in the DataFrame
print("Number of rows in DataFrame are: " + str(hygiene_20_df['BusinessName'].count()))
print("-"*50)

# Display the first 10 rows of the DataFrame
hygiene_20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query_rating = {'LocalAuthorityName':{'$regex':'London'}, 
                'RatingValue':{'$gte':4}
               }

# Use count_documents to display the number of documents in the result
rating_gte_4 = establishments.count_documents(query_rating)
print("There are " + str(rating_gte_4) + " documents for this query")
print("-"*50)

#Save the results of the query to variable
rating_gte_4_list = list(establishments.find(query_rating))

# Display the first document in the results using pprint
rating_gte_4_list[0]
```

```python
# Convert the result to a Pandas DataFrame
rating_gte_4_df = pd.DataFrame(rating_gte_4_list)

# Display the number of rows in the DataFrame
print("Number of rows in DataFrame are: " + str(len(rating_gte_4_df)))
print("-"*50)

# Display the first 10 rows of the DataFrame
rating_gte_4_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
print("Penang Flavours GeoCode Information")
print(establishments.find_one({"BusinessName":"Penang Flavours"},{'geocode'}))
print("-"*60)

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

#Creating variables to take into account the min/max of Longitude and Latitude
min_lat = latitude - degree_search
max_lat = latitude + degree_search
min_lon = longitude - degree_search
max_lon = longitude + degree_search

query = {'geocode.latitude':{'$gte': min_lat, '$lte': max_lat},
         'geocode.latitude':{'$gte': min_lat, '$lte': max_lat},
                'RatingValue':5
        }

sort =  [('score.hygiene',1)]

good_businesses_near_Penang = list(establishments.find(query).sort(sort))

# Print the results
for i in range(4):
    pprint(good_businesses_near_Penang)
```

```python
# Convert result to Pandas DataFrame
good_businesses_near_Penang_db = pd.DataFrame(good_businesses_near_Penang)
good_businesses_near_Penang_db.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match_query = {'$match': {'scores.Hygiene': 0}}

groups = {'$group': {'_id':'$LocalAuthorityName',
                     'count': {'$sum':1}
                    }
         }

sort = {'$sort' : {'count': -1}}

pipeline = [match_query,groups,sort]

# Print the number of documents in the result
results = list(establishments.aggregate(pipeline))
print("Number of documents in the query: " + str(len(results)))
               
# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
pipeline_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print("Number of rows in DataFrame are: " + str(len(pipeline_df)))
print("-"*50)

# Display the first 10 rows of the DataFrame
pipeline_df.head(10)
```
